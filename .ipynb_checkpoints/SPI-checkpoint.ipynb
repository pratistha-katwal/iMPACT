```python
dataset= xr.open_dataset('/Applications/MAMP/htdocs/Plotly_Visualization_ppt/Dataset/dataset_SeasonalYear_updated.nc')
dataset=dataset.rio.write_crs('EPSG:4326')

shp = gpd.read_file('/Applications/MAMP/htdocs/Plotly_Visualization_ppt/Shapefile/Nepal_bnd_WGS84.shp')
shp= shp.to_crs(dataset.rio.crs)
```

```python
monthly= dataset['tp'].resample(time='ME').sum(dim=['time'])


```

```python
#getting the numpy array

precip_np= monthly.values
precip_np= np.nan_to_num(precip_np, nan=0.0)
times= monthly.time.values
lat= monthly.lat.values
lon= monthly.lon.values

precip_np.shape
```

```python
#Parameter

scale= 3
distribution= indices.Distribution.gamma 
periodicity= compute.Periodicity.monthly
data_start_year= monthly.time.dt.year[0].item()
data_end_year= monthly.time.dt.year[-1].item()

#spi initialization
spi3= np.zeros_like(precip_np)


#Computing spi for each frid

for ilat in range(precip_np.shape[1]):
    for ilon in range(precip_np.shape[2]):
        #getting the time series
        precip_series=precip_np[:,ilat,ilon]
        
        #calculating spi3
        
        spi3[:,ilat,ilon]= indices.spi(
        precip_series,
        scale=3,
        distribution= distribution,
        periodicity= periodicity,
        data_start_year=data_start_year,
        calibration_year_initial= data_start_year,
        calibration_year_final= data_end_year)



```

```python
#converting back to x array
spi_da = xr.DataArray(
    spi3,
    dims=['time', 'lat', 'lon'],
    coords={
        'time': times,
        'lat': lat,
        'lon': lon,
    },
    name='spi3'
)

spi_da = spi_da.rio.write_crs('EPSG:4326')
```

```python
spi_december_2024 = spi_da.sel(time='2024-10')


spi_december_2024.plot(cmap='RdBu', vmin=-2, vmax=2)
plt.title('SPI-3 Map for December 2024')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs  # For geographic maps

# Select SPI-3 for December 2024
spi_december_2024 = spi_da.sel(time='2024-12')

# Create a figure with a geographic projection
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())  # Use PlateCarree for lat/lon

# Plot SPI with improved settings
plot = spi_december_2024.plot(
    ax=ax,
    cmap='RdBu',  # Red-Blue colormap (dry vs wet)
    vmin=-2.5,    # Minimum SPI value (extreme drought)
    vmax=2.5,     # Maximum SPI value (extreme wet)
    levels=11,    # Number of color bins
    extend='both', # Extend colors for values beyond vmin/vmax
    transform=ccrs.PlateCarree()  # Coordinate system
)

# Add geographic features
ax.coastlines(linewidth=0.5)
ax.gridlines(draw_labels=True)  # Add latitude/longitude grid

# Customize colorbar
cbar = plot.colorbar
cbar.set_label('SPI-3 Value', rotation=270, labelpad=15)
cbar.ax.set_yticklabels([
    'Extreme Drought', 'Severe Drought', 'Moderate Drought',
    'Mild Drought', 'Normal', 'Mild Wet', 'Moderate Wet',
    'Severe Wet', 'Extreme Wet'
])

# Add title and show
plt.title('SPI-3 for December 2024 (3-Month Drought Indicator)')
plt.show()
```
